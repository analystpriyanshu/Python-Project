```python
#DATA COLLECTION:

df=pd.read_csv('datasets/household_power_consumption.csv')
df
```

```python
df.info()
# its give information of all columns of datasets...
```

```python
df.describe()
#it gives descriptive / statistic value of given data set 
```

```python
#DATA CLEANING:
'''Now we change the column type of date from object to datetime type....'''
df['Date'] = pd.to_datetime(df['Date'] , format= '%d/%m/%Y')

df.dtypes
```

```python
df['Time']=pd.to_datetime(df["Time"])
```

```python
df['Time']=df['Time'].dt.strftime('%H:%M:%S')
df
```

```python
df.dtypes
```

```python
'''Split the date column into day, month, and year'''
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df
```

```python
'''Split the time column into hour and minute'''
#tushar singhal
df[['Hour' , 'Minute','Second']] = df['Time'].str.split(':' , expand= True)
```

```python
df['Hour']=df['Hour'].astype(int)
df['Minute']=df['Minute'].astype(int)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df_sam=df.sample(100000)
df_sam.head()
```

```python
df_sam['Global_active_power']=pd.to_numeric(df_sam['Global_active_power'] , errors='coerce')
df_sam['Global_reactive_power']=pd.to_numeric(df_sam['Global_reactive_power'] , errors='coerce')
df_sam['Global_intensity']=pd.to_numeric(df_sam['Global_intensity'] , errors='coerce')
df_sam['Voltage']=pd.to_numeric(df_sam['Voltage'] , errors='coerce')
df_sam['Sub_metering_1']=pd.to_numeric(df_sam['Sub_metering_1'] , errors='coerce')
df_sam['Sub_metering_2']=pd.to_numeric(df_sam['Sub_metering_2'] , errors='coerce')
df_sam['Sub_metering_3']=pd.to_numeric(df_sam['Sub_metering_3'] , errors='coerce')
```

```python
'''Now fill the errors with the help of mean...'''
df_sam['Global_active_power'].fillna(df_sam['Global_active_power'].mean(), inplace=True)
df_sam['Global_reactive_power'].fillna(df_sam['Global_reactive_power'].mean(), inplace=True)
df_sam['Global_intensity'].fillna(df_sam['Global_intensity'].mean(), inplace=True)
df_sam['Voltage'].fillna(df_sam['Voltage'].mean(), inplace=True)
df_sam['Sub_metering_1'].fillna(df_sam['Sub_metering_1'].mean(), inplace=True)
df_sam['Sub_metering_2'].fillna(df_sam['Sub_metering_2'].mean(), inplace=True)
df_sam['Sub_metering_3'].fillna(df_sam['Sub_metering_3'].mean(), inplace=True)
df_sam.head()
```

```python
'''Now check that the eroors are replace by mean value or not...'''
df_sam.isnull().sum()
```

```python
'''Now creatin a column for total sub_metering....'''

df_sam['Total_Metering']= df_sam['Sub_metering_1'] + df_sam['Sub_metering_2'] +df_sam['Sub_metering_3']
df_sam.head()
```

```python
#Exploratory Data Analysis (EDA):
'''Year wise total power consumption....'''
plt.figure(figsize=(20,10))
plt.suptitle('Yearly-Total watt/hour power consumption', fontsize=20,y=1)
df_sam.groupby(df_sam.Year)['Global_active_power'].sum().plot(kind='bar',xlabel='Year',ylabel='Readings in watt/hour')
plt.show()
```

```python
'''Year wise total metering....'''
plt.figure(figsize=(20,10))
plt.suptitle('Yearly-Total Metering', fontsize=20,y=1)
df_sam.groupby(df_sam.Year)['Total_Metering'].sum().plot(kind='bar',xlabel='Year',ylabel='Readings in watt/hour')
plt.show()
```

```python
'''Year wise total power consumption....'''
plt.figure(figsize=(20,10))
plt.suptitle('Monthly Voltage', fontsize=20,y=1)
df_sam.groupby(df_sam.Month)['Voltage'].sum().plot(kind='bar',xlabel='Month',ylabel='Readings in watt/hour')
plt.show()
```

```python

plt.figure(figsize=(20,10))
sns.lineplot(x='Hour',y='Total_Metering',data=df_sam,color='yellow')
list=np.arange(0,26,2)
plt.xticks(list)
plt.show()
```

```python
#Here it can be seen that in Hourly vs Total_metering...
#The total metering increases after 6 and its less during 0 to 5 hours...
```

```python

plt.figure(figsize=(20,10))
sns.lineplot(x='Month',y='Total_Metering',data=df_sam,color='yellow')
list=np.arange(0,12,1)
plt.xticks(list)
plt.show()
```

```python
#Here it can be seen that in Monthly vs Total_metering...
#The total metering decreases after 6 month and its highest in first month and lowest in eight month...
'''tushar singhal'''
```

```python
# Plotting the pie chart
plt.figure(figsize=(8, 6))
month_labels=['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
plt.pie(x=df_sam.groupby(df_sam['Month'])['Total_Metering'].sum(), autopct='%1.1f%%', startangle=0, labels= month_labels)
plt.title('Pie Chart of total metering by month')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
